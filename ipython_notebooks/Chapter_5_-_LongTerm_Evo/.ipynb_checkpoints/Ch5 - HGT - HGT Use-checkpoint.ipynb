```python
def add_treatment(dic, series, name, seed, longname=None, shortid=None, 
                  factors={}, skip_seeds=None,
                  location=None, filename=None):
    shortlabel = name
    if series:
        shortlabel = series + "_" + name
        
    if not longname:
        longname = "\n".join( name.split("_") )
                        
    if not location:        
        location = series + "_" + name + "_" + seed
    
    if skip_seeds == None:
        skip_seeds = []
    
    dic[shortlabel] = {'longlabel': longname, 'shortlabel': shortlabel, 'series': series, 'id': shortid,
                  'dir': location, 'filename': filename, 'factors': factors, 'skip': skip_seeds} 
```

#### Dataset Manipulation (Grab/Thin)

```python
def grab_set(ts, factor=None, series=None):
    if not factor and not series:
        return ts
    if series and factor:
        thing = [ts[t]['shortlabel'] for t in ts if 
               (factor in ts[t]['factors'].values() and 
               ts[t]['series'] == series)]
        
        #print "HI", series, factor
        #print thing
        
    elif factor:
        thing = [ts[t]['shortlabel'] for t in ts if 
               factor in ts[t]['factors'].values()]
    elif series:
        thing = [ts[t]['shortlabel'] for t in ts if 
               ts[t]['series'] == series]
        
    subset = collections.OrderedDict()
    for t in thing:
        subset[t] = ts[t]
        
    return subset

def test_grab_set(treatments):

    print "Subsetting"
    print treatments.keys()
    print
    print "Subset, selecting on series AC-1"
    print grab_set(series="AC-1", ts=treatments).keys()
```

```python
## THIN SETS
def thin_set(ts, factor=None, series=None):

    thing = ts.keys()
    
    if factor:
        for t in ts.keys():
            if factor in ts[t]['factors'].values():
                thing.remove(t)
                
    if series:
        for t in ts.keys():
            if series == ts[t]['series']:
                thing.remove(t)
        
    subset = collections.OrderedDict()
    for t in thing:
        subset[t] = ts[t]
        
    return subset

def test_thin_set(treatments):
    print
    print "Thinning Set"
    print treatments.keys()
    thingy = grab_set(series="AC-2", factor="Homol", ts=treatments)
    print
    print "Subset selecting by HOMOL factor"
    print thingy.keys()
    print
    print "Thinned by removing MRI factor"
    print thin_set(factor="MRI", ts=thingy).keys()
```

```python
%pwd 

def generate_data(use_pickled=False):
    picklename = "task_performance.pd.pickle"

    ## check for pickling!
    if use_pickled and os.path.exists(picklename):
        task_performance_data = pd.read_pickle(picklename)
        return task_performance_data
        
    task_performance_data = []
    for t in treatments:  

        print t

        task_performance = []
        task_performance_persistent = []

        for i in range(1,51):

            bit = str(i).zfill(2)
            globstring = '../*' + treatments[t]['dir'] + "*" + bit + '/data/tasks.dat*'
            dfiles = sorted(glob.glob(globstring))
            #print globstring

            for item in dfiles:
                sys.stdout.write('.')  
                #sys
                #print item
                df = pd.read_csv(item, header=None, comment="#", 
                                  delimiter=' ', usecols=range(78),
                                  skipinitialspace=True) #, names=fields)


                task_performance.append( len(df.tail(2000).mean().nonzero()[0]) - 3 )
                # 3.6 is 1% of the population at any given time
                task_performance_persistent.append( len(df.tail(2000).mean().where(lambda x : x > 3.6).dropna().nonzero()[0]) - 3 )
        print " done"        

        intermediate = pd.DataFrame()
        intermediate['tasks performed'] = task_performance
        intermediate['label'] = [t] * len(task_performance)
        intermediate['persistence'] = ['incl. transient'] * len(task_performance)
        intermediate['Treatment'] = [ treatments[t]['longlabel'] ] * len(task_performance)

        task_performance_data.append(intermediate)

        intermediate = pd.DataFrame()
        intermediate['tasks performed'] = task_performance_persistent
        intermediate['label'] = [t] * len(task_performance_persistent)
        intermediate['persistence'] = ['persistent'] * len(task_performance_persistent)
        intermediate['Treatment'] = [ treatments[t]['longlabel'] ] * len(task_performance_persistent)

        task_performance_data.append(intermediate)

    task_performance_data = pd.concat(task_performance_data)
    
    ## if we generate the data, pickle it
    task_performance_data.to_pickle("task_performance.pd.pickle")

    return task_performance_data

task_performance = generate_data(use_pickled=True)     
print task_performance.head()
```
